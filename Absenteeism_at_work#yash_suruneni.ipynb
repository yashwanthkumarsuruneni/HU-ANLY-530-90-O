```python
absent_records.drop(['Month of absence','Day of the week','Seasons','Weight','Height','Hit target'], 1, inplace = True)
```

```python
absent_records['Age Group'] = ['young employee' if 25<=age<=35 else 'mid age employee' if 36<=age<=45 else 'old employee' if age>=45 else None\
                      for age in list(absent_records['Age'].replace("R", 0).astype('int64').values)  ] 

absent_records['smoke_cat'] = ['no' if period<=0 else 'yes' if period >=1 else None \
                               for period in list (absent_records['Social smoker'].values)]

absent_records['Disciplinary action'] = ['not taken' if period<=0 else 'taken' if period >=1 else None \
                               for period in list (absent_records['Disciplinary failure'].values)]

absent_records['drink_cat'] = ['no' if period<=0 else 'yes' if period >=1 else None \
                               for period in list (absent_records['Social drinker'].values)]

absent_records['Education_level'] = ['high school' if  period ==1 else 'graduate' if  period ==2 else "postgraduate" if  period ==3 else 'doctor' if  period ==4 else  None \
                               for period in list (absent_records['Education'].values)]

absent_records['Transportation costs'] = ['cheap' if 100<=fare<=200 else 'expensive' if 200<=fare<=300 else 'very expensive' if fare>300 else None \
                                   for fare in list(absent_records["Transportation expense"].values)]

absent_records['distance'] = ['close' if 0<=dist<=15 else 'far' if 15<=dist<=35 else 'very far' if dist>35 else None \
                                   for dist in list(absent_records["Distance from Residence to Work"].values)]

absent_records['absenteeism category'] = ['group_0' if period==0 else 'group_1' if 0 <=period<=6 else 'group_2' if period>=6 else None \
                               for period in list (absent_records["Absenteeism time in hours"].values)]

```

```python
absent_records = pd.DataFrame(data = absent_records).dropna()

absent_records.isnull().sum()
```

```python
# dividing into groups

import warnings
warnings.filterwarnings("ignore")
import numpy as np

group_0 = absent_records['Absenteeism time in hours'] == 0
x = absent_records['Absenteeism time in hours'] > 0
y = absent_records['Absenteeism time in hours'] <= 6 
group_1 = np.logical_and(x, y)
group_2 = absent_records['Absenteeism time in hours'] > 6

absent_records.loc[group_0,'Absenteeism time in hours'] = 0
absent_records.loc[group_1,'Absenteeism time in hours'] = 1
absent_records.loc[group_2,'Absenteeism time in hours'] = 2
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
hrs_label = pd.Series(label_encoder.fit_transform(absent_records['Absenteeism time in hours']>6))
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import cross_val_score
from sklearn import metrics

absent_records['Age'] = absent_records['Age'].replace("R", 0).astype('int64')
X_pd = pd.DataFrame(data = absent_records['Work load Average/day ']) 
X_replace = X_pd.replace(',','', regex=True)
X_np = X_replace.values
X_fa = X_np.astype('float64')
absent_records["Work load Average/day "] = X_fa


y = absent_records['Absenteeism time in hours']
X = absent_records.drop(['Absenteeism time in hours','Weight'],axis=1)

# split dataset to 70-30 training and testing resp.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 ,random_state=52)

#Shape of  training and test datasets
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier 
model = tree.DecisionTreeClassifier() 
model = model.fit(X_train, y_train)
```

```python
# Evaluating model performance 

from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score
import seaborn as sea
import matplotlib.pyplot as plt

y_predict = model.predict(X_test)

def conftable(test,pred, imagename):
    confmatrix= confusion_matrix(y_test, y_predict)
    plt.matshow(confmatrix)
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('Truth')
    plt.xlabel('Predicted')
    plt.savefig(imagename)
    plt.show()
    print(confmatrix)
conftable(y_test,y_predict,"conf")

print(accuracy_score(y_test, y_predict)*100)
```

## Random Forest Classifiers

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)

def conftable(test,pred, imagename):
    confmatrix= confusion_matrix(y_test, y_predict)
    plt.matshow(confmatrix)
    plt.title('Confusion matrix')
    plt.colorbar()
    plt.ylabel('Truth')
    plt.xlabel('Predicted')
    plt.savefig(imagename)
    plt.show()
    print(confmatrix)
conftable(y_test,y_predict,"conf")

print(accuracy_score(y_test,y_predict)*100)
```
